Ejercicio 1: Saludo Personalizado

        Crea un programa que pida al usuario su nombre y edad, y luego imprima un saludo personalizado.
        Usa input para capturar el nombre y la edad.
        Usa print para mostrar el saludo.

```python
while(True):
  nombre = str(input("Cual es tu nombre? "))
  if not nombre.isalpha():
    print("Por favor, ingresa un nombre válido.")
  else:
    break

while(True):
  edad = input("Cual es tu edad? ")
  if not edad.isdigit():
    print("Por favor, ingresa una edad válida.")
  else:
    break

print(f"Buenas buenas {nombre}, deja que adivino tu edad: {edad} :D")

```

Ejercicio 2: Manipulación de Cadenas

    Crea una cadena con tu nombre completo.
    Usa los métodos de cadenas para:
        Convertir el nombre a mayúsculas y minúsculas.
        Contar el número de letras en tu nombre.
        Reemplazar tu primer nombre por otro nombre.



```python
nombre = "Sebastian Ignacio Antón Carmona"
print("nombre en mayúscula: ", nombre.upper())
print("nombre en minúscula: ", nombre.lower())
print("cantidad de caracteres del nombre", len(nombre))

nombre2 = nombre.replace("Sebastian", "Juan")
print("primer nombre reemplazado", nombre2)
```


Ejercicio 3: Lista de Nombres

    Pide al usuario que ingrese una lista de nombres separados por comas.
    Usa split para convertir esta cadena en una lista de nombres.
    Imprime cada nombre en una línea separada.

```python

nombres = str(input("Ingresa nombres separados por comas: "))

listaNombres = nombres.split(",")
for indice, nombre in enumerate(listaNombres):
  nombre = nombre.strip()
  print(f"Nombre {indice+1}: {nombre} \n")
```
